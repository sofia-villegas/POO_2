```python
import sqlite3
from sqlite3 import Error

def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()

if __name__ == '__main__':
    create_connection(r"C:\sqlite\db\pythonsqlite.db")
```

Debes conectarte a la base de datos pasando el archivo de la base de datos a la función connect(). Si la base de datos no existe, se crea una nueva base de datos y te conectas a ella. Y en caso de que la base de datos exista, te conectas a esa base de datos.
